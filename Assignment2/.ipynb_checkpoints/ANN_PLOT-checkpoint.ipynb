```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from operator import itemgetter
```

```python
def plot_curve(Xdata,Ydatas,datalabels,title,filename,Xlabel='Iterations',Ylabel='Accuracy'):
    
    for i in range(len(Ydatas)):
        
        plt.plot(Xdata,
                 Ydatas[i],
                 label=datalabels[i])
    
    filename = './ANN/' + filename
    plt.xlabel(Xlabel)
    plt.ylabel(Ylabel)
    plt.title(title)
    plt.legend()
    plt.savefig(filename)
    plt.show()
```

```python
bp = pd.read_csv('./ANN/BP/BACKPROP_LOG.csv')
Xdata = bp['iteration']

Ydatas = [bp['MSE_tst'],bp['MSE_trg']]
datalabels = ['MSE_test','MSE_train']
title = 'Learning Error Curve Back Propagation'
filename = 'BP_PLOT_ERROR.png'
plot_curve(Xdata,Ydatas,datalabels,title,filename,Ylabel='Error')

Ydatas = [bp['acc_trg'],bp['acc_tst']]
datalabels = ['acc_train','acc_test',]
title = 'Learning Curve of Back Propagation'
filename = 'BP_PLOT_ACCURACY.png'
plot_curve(Xdata,Ydatas,datalabels,title,filename)

print("test accuracy: %s, time %s"%(bp['acc_tst'].iloc[-1],bp['elapsed'].iloc[-1]))
```

```python
SA_Records = dict()
Accu_ALL = dict()
for CE in [0.15,0.35,0.55,0.75,0.95]:
    for  T in [1e8,1e10,1e12]:
        if T == 1e12:
            oFile = "./ANN/SA/%s_%s_LOG.csv"%(CE,'1e+12')
        else:
            oFile = "./ANN/SA/%s_%s_LOG.csv"%(CE,T)
        sa = pd.read_csv(oFile)
        last_row = list()
        for key in list(sa.columns.values): 
            last_row.append(sa[key].iloc[-1])
        SA_Records[(CE,T)] = last_row
        Accu_ALL[(CE,T)] = last_row[4]
A = sorted((value, key) for (key,value) in Accu_ALL.items())

for a,p in A:
    print ("CE: %s T: %s Accuracy: %s time: %s"%(p[0],p[1],a,SA_Records[p][-1]))
```

```python
Xdata = [0.15,0.35,0.55,0.75,0.95]
Ydatas = list()
for T in [1e8,1e10,1e12]:
    y = list()
    for CE in [0.15,0.35,0.55,0.75,0.95]:
        y.append(Accu_ALL[(CE,T)])
    Ydatas.append(y)
datalabels =  ["T:1e8","T:1e10","T:1e12"]
title = 'SA: Accuracy Achieved by Different CE & T'
filename = 'SA_PLOT_CET.png'
plot_curve(Xdata,Ydatas,datalabels,title,filename,Xlabel="CE")
```

```python
for T in [1e8,1e10,1e12]:
    Ydatas = list()
    datalabels = list()
    for CE in [0.15,0.35,0.55,0.75,0.95]:
        if T == 1e12:
            oFile = "./ANN/SA/%s_%s_LOG.csv"%(CE,'1e+12')
        else:
            oFile = "./ANN/SA/%s_%s_LOG.csv"%(CE,T)
        sa = pd.read_csv(oFile)
        Ydatas.append(sa['acc_tst'])
        datalabels.append("CE: %s"%(CE))
    plt.rcParams['figure.figsize'] = (6.0, 4.0)
    plot_curve(sa['iteration'],Ydatas,datalabels,"SA: learning curve for different CE (T = %s)"%(T),"SA_%s.png"%(T),Xlabel='Iterations',Ylabel='Accuracy')
```
